```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('/content/winequality.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['chlorides']=df['chlorides'].fillna(df['chlorides'].median())
```

```python
df['pH']=df['pH'].fillna(df['pH'].median())
```

```python
df['sulphates']=df['sulphates'].fillna(df['sulphates'].median())
df.isnull().sum()
```

```python
import re as I
q1 = df['quality'].quantile(0.25)
q3 = df['quality'].quantile(0.75)
iqr = q3 - q1

upper = q3 + 3 * iqr
lower = q1 - 3 * iqr

df = df[(df['quality'] >= lower) & (df['quality'] <= upper)]
df.shape
```

```python
x=df.drop(columns='quality')
y=df['quality']
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
x_std=scaler.transform(x)
```

```python
x=x_std
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

print(x.shape,x_train.shape,x_test.shape)
```

```python
cf=RandomForestRegressor(n_estimators=100,max_depth=4,random_state=42)
```

```python
cf.fit(x_train,y_train)
```

```python
x_train_pred=cf.predict(x_train)
```

```python
x_train_r=r2_score(y_train,x_train_pred)
```

```python
print(x_train_r)
```

```python
x_test_pred=cf.predict(x_test)
x_test_r=r2_score(y_test,x_test_pred)
print(x_test_r)
```
