```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df=pd.read_csv('/content/heart_disease.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
df['trestbps'] = df['trestbps'].fillna(df['trestbps'].median())

df['chol'] = df['chol'].fillna(df['chol'].median())

df['thalach'] = df['thalach'].fillna(df['thalach'].median())
```

```python
le = LabelEncoder()

df['sex'] = le.fit_transform(df['sex'])
```

```python
df['target'].value_counts()
```

```python
df.groupby('target').mean()
```

```python
x=df.drop(columns='target',axis=1)
y=df['target']
```

```python
print(x)
```

```python
print(y)
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(x)
x_std=scaler.transform(x)
```

```python

x=x_std
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
cf=svm.SVC(kernel="linear")
```

```python
cf.fit(x_train,y_train)
```

```python
x_train_pred=cf.predict(x_train)


```

```python
x_train_acc=accuracy_score(x_train_pred,y_train)
```

```python
print(x_train_acc)
```

```python
x_test_pred=cf.predict(x_test)
```

```python
x_test_acc=accuracy_score(x_test_pred,y_test)
```

```python
print(x_test_acc)
```
